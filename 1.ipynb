```python
text_columns = ["Packed Cell Volume","White Blood Cell Count","Red Blood Cell Count"]
for i in text_columns:
    print(f"{i}:{data[i].dtype}")
```

```python
def convert_to_numeric(data,column):
    data[column] = pd.to_numeric(data[column],errors="coerce")

for column in text_columns:
    convert_to_numeric(data,column)

for i in text_columns:
    print(f"{i}:{data[i].dtype}")
```

```python
missing = data.isnull().sum().sort_values(ascending=False)
missing
```

```python
def num_fill(data,column):

    mean_val = data[column].mean()
    data[column] = data[column].fillna(mean_val)

def obj_fill(data,column):

    mode_val = data[column].mode()[0]
    data[column] = data[column].fillna(mode_val)
```

```python
num_val = [i for i in data.columns if data[i].dtype!='object']

obj_val = [i for i in data.columns if data[i].dtype=='object']

for i in num_val:
    num_fill(data,i)

for i in obj_val:
    obj_fill(data,i)
```

```python
# Red Blood Cells
data['Red Blood Cells'] = data['Red Blood Cells'].map({
    'normal': 1,
    'abnormal': 0
})

# Pus Cell
data['Pus Cell'] = data['Pus Cell'].map({
    'normal': 1,
    'abnormal': 0
})

# Pus Cell Clumps
data['Pus Cell Clumps'] = data['Pus Cell Clumps'].map({
    'present': 1,
    'notpresent': 0
})

# Bacteria
data['Bacteria'] = data['Bacteria'].map({
    'present': 1,
    'notpresent': 0
})

# Hypertension
data['Hypertension'] = data['Hypertension'].map({
    'yes': 1,
    'no': 0
})

# Diabetes Mellitus
data['Diabetes Mellitus'] = data['Diabetes Mellitus'].map({
    'yes': 1,
    'no': 0
})

# Coronary Artery Disease
data['Coronary Artery Disease'] = data['Coronary Artery Disease'].map({
    'yes': 1,
    'no': 0
})

# Appetite
data['Appetite'] = data['Appetite'].map({
    'good': 1,
    'poor': 0
})

# Pedal Edema
data['Pedal Edema'] = data['Pedal Edema'].map({
    'yes': 1,
    'no': 0
})

# Anemia
data['Anemia'] = data['Anemia'].map({
    'yes': 1,
    'no': 0
})

# Target Column
data['Classification'] = data['Classification'].map({
    'ckd': 1,
    'notckd': 0
})
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

models = []

models.append(("Naive Bayes", GaussianNB()))
models.append(("KNN", KNeighborsClassifier(n_neighbors=8)))
models.append(("Random Forest", RandomForestClassifier()))
models.append(("Decision Tree", DecisionTreeClassifier()))
models.append(("SVM", SVC(kernel='linear')))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

for name, model in models:

    print(name, model)
    print()

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(confusion_matrix(y_test, y_pred))
    print("\n")

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("\n")

    print("Precision:", precision_score(y_test, y_pred))
    print("\n")

    print("Recall:", recall_score(y_test, y_pred))
    print("\n")

    print("F1 Score:", f1_score(y_test, y_pred))
    print("\n")
```
